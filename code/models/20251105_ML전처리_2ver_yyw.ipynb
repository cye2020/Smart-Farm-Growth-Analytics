```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

def regression_metrics(y_true, y_pred, n_features):
    """

    Parameters:
    - y_true: 실제 값 (array-like)
    - y_pred: 예측 값 (array-like)
    - n_features: 입력 변수(특성) 수 (int)

    Returns:
    - dict: MSE, MAE, RMSE, R2, Adjusted R2
    """
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    n = len(y_true)
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - n_features - 1)

    return {
        "MSE": mse,
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2, # 1-어쩌구저쩌구(성능을저하시키는지표) = 전체 성능에서 얼마나 방해가 있었는지 알려주는 지표. 얘가 마이너스면? 모델이 동전던지기보다 못 맞힌다.
        "Adjusted R2": adj_r2
    }
```

```python
xgb_model = XGBRegressor(
    objective='reg:squarederror',
    tree_method="hist",
    device="cuda",
    random_state=42)

param_grid = {
    'n_estimators': [200, 300],
    'max_depth': [2, 3, 4],
    'learning_rate': [0.01, 0.05, 0.1, 0.2]}
```

```python
grid_search1 = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    cv=3,
    scoring='neg_mean_absolute_error',
    verbose=1,
    n_jobs=-1
)
```

```python
grid_search1.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid_search1.best_params_)
print("Best Score (Negative MAE):", grid_search1.best_score_)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from tqdm import tqdm

# 파이프라인 정의: 결측치 처리 → 표준화 → 랜덤포레스트 회귀
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median', add_indicator=True)),
    ('scaler', StandardScaler()),
    ('rf', RandomForestRegressor(random_state=42))
])

# 탐색할 파라미터 설정 (rf 단계에 맞춰 prefix 붙이기)
param_grid = {
    'rf__n_estimators': [200, 300, 500],
    'rf__max_depth': [5, 10, None],
    'rf__min_samples_split': [2, 5],
    'rf__min_samples_leaf': [1, 2, 5],
    'rf__max_features': ['sqrt', 0.7, None]
}

# GridSearchCV 설정
grid_search_rf = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    scoring='r2',   # 회귀 문제이므로 R² 사용
    cv=3,
    verbose=2,
    n_jobs=-1
)

# 학습
grid_search_rf.fit(X_train, y_train)

# 결과 확인
print("Best Params:", grid_search_rf.best_params_)
print("Best Score (R²):", grid_search_rf.best_score_)
```

```python
from lightgbm import LGBMRegressor
from sklearn.model_selection import GridSearchCV

# 일반 회귀 모델 (R² 평가 가능)
model = LGBMRegressor(
    objective='regression',   # 평균값 예측
    random_state=42
)

# 하이퍼파라미터 후보
param_grid = {
    'n_estimators': [500, 1000, 2000],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 5, 8],
    'num_leaves': [31, 63]
}

# GridSearchCV로 최적 파라미터 탐색
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='r2',   # R² 기준으로 평가
    cv=3,
    verbose=1,
    n_jobs=-1
)
```

```python
from sklearn.metrics import r2_score

# 1. 학습 실행
grid_search.fit(X_train, y_train)

# 2. 최적 파라미터 확인
print("Best Parameters:", grid_search.best_params_)
```

```python
# 3. 최적 모델 가져오기
best_model = grid_search.best_estimator_

# 4. 학습 데이터 혹은 검증 데이터로 예측
y_pred = best_model.predict(X_test)

# 5. R² 출력
r2 = r2_score(y_test, y_pred)
print("R² Score:", r2)
```
